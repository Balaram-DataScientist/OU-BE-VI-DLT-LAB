# 1.Merge
Merging means combining multiple tensors into one tensor in a certain
dimension.
Tensors can be merged using concatenate and stack operations.
The concatenate operation does not generate new dimensions. It only merges along existing dimensions.
The stack operation creates new
dimensions.
Whether to use the concatenate or stack operation to merge tensors depends on whether a new dimension needs to be created for a
specific scene.

```python
# Import tf
import tensorflow as tf
```

```python
a = tf.random.normal([4,35,8]) # Create gradebook A
b = tf.random.normal([6,35,8]) # Create gradebook B
tf.concat([a,b],axis=0) # Merge gradebooks
```

# a.Concatenation

```python
# Illegal merge.
a = tf.random.normal([4,32,8])
b = tf.random.normal([6,35,8])
tf.concat([a,b],axis=0)   # Illegal merge. Second dimension is different.
```

```python
a = tf.random.normal([10,35,4])
b = tf.random.normal([10,35,4])
tf.concat([a,b],axis=2) # Merge along the last dimension
```

```python
a = tf.random.normal([35,8])
b = tf.random.normal([35,8])
tf.concat([a,b],axis=0) # No class dimension
```

# b.Stack

```python
a = tf.random.normal([35,8])
b = tf.random.normal([35,8])
tf.stack([a,b],axis=0) # Stack a and b and insert new dimension
```

```python
a = tf.random.normal([35,8])
b = tf.random.normal([35,8])
tf.stack([a,b],axis=-1) # Insert new dimension at the end
```

```python
# Illegal
a = tf.random.normal([35,4])
b = tf.random.normal([35,8])
tf.stack([a,b],axis=-1) # Illegal use of stack function.
```

# 2.Split
The inverse process of the merge operation is split, which splits a tensor
into multiple tensors.

```python
x = tf.random.normal([10,35,8])
# Cut into 10 pieces with equal length
result = tf.split(x, num_or_size_splits=10, axis=0)
result
```

```python
len(result) # Return a list with 10 tensors of equal length
```

```python
result[0]
```

```python
x = tf.random.normal([10,35,8]) # Split tensor into 4 parts
result = tf.split(x, num_or_size_splits=[4,2,2,2] ,axis=0)
len(result)
```

```python
"""to divide one certain dimension by a length
of 1, we can use the tf.unstack(x, axis) function. This method is a special
case of tf.split. The splitting length is fixed as 1."""
x = tf.random.normal([10,35,8])
result = tf.unstack(x,axis=0)
len(result) # Return a list with 10 tensors
```

# 3. Statistics
In neural network  various statistical attributes need to be computed, such as maximum, minimum, mean, and norm.
Because tensors usually contain a lot of data, it is easier to infer the distribution of
tensor values by obtaining the statistical information of these tensors.

# a.Norm
Norm is a measure of the length of a vector. It can be generalized to
tensors.
there 3 types of Norms
L1 Norm ,L2 Norm and infinite Norm

```python
x = tf.ones([2,2])
tf.norm(x,ord=1) # L1 norm
```

```python
tf.norm(x,ord=2) # L2 norm
```

```python
import numpy as np
tf.norm(x,ord=np.inf) #  norm
```

# b.Max, Min, Mean, and Sum

```python
#the maximum, minimum, mean, and sum of tensors in a certain dimension or in all dimensions.
x = tf.random.normal([4,10])
x
```

```python
tf.reduce_max(x,axis=1) # get maximum value at 2nd dimension
```

```python
tf.reduce_min(x,axis=1) # get the minimum value at 2nd dimension
```

```python
# Find the mean probabilities of each sample:
tf.reduce_mean(x,axis=1)
```

```python
x = tf.random.normal([4,10])
x
```

```python
tf.reduce_max(x),tf.reduce_min(x),tf.reduce_mean(x)
```

# c. Error Function

```python
""" error function is the  error of each sample can be obtained through
the MSE function, and the average error of the sample  """
out = tf.random.normal([4,10]) # Simulate output
print("out=",out)

y = tf.constant([1,2,2,0]) # Real labels
print("\n 1. y with real labels=",y)
y = tf.one_hot(y,depth=10) # One-hot encoding
print("\n 2. y with one hot encoding=",y)
loss = tf.keras.losses.mse(y,out) # Calculate loss of each sample
print("\n 3. mse loss =",loss)
loss = tf.reduce_mean(loss) # Calculate mean loss
print("\n 4. reduced mean loss=",loss)
```
